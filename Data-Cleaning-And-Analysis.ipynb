```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
fatalities = pd.read_csv('fatalities.csv')
fatalities['date_of_death'] = pd.to_datetime(fatalities['date_of_death'])
fatalities['date_of_event'] = pd.to_datetime(fatalities['date_of_event'])
fatalities
```

```python
#fatalities over the year
#Overall fatalitites
yearly_fatalities = fatalities['date_of_death']

death_counts = yearly_fatalities.dt.year.value_counts().sort_index()
print('Overall deaths:', len(fatalities))
plt.figure(figsize=(10, 6))
death_counts.plot(kind='bar')
plt.title('Overall Death Counts')
plt.xlabel('Date')
plt.ylabel('Count')
plt.show()
```

```python
#Region based on violence
location_based = fatalities.event_location_region.value_counts()
df_region = pd.DataFrame({'area': location_based.index, 'count': location_based.values})

plt.style.use('_mpl-gallery-nogrid')

colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(location_based)))

# plot
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(df_region['count'], colors=colors, 
                                  autopct='%1.1f%%', radius=1, center=(0, 0),
                                  wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

ax.legend(wedges, df_region['area'], title='Areas', loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
ax.axis('equal')  

plt.show()
```

```python
# District Based Analysis on Gaza Strip
district_based = fatalities[fatalities['event_location_region'] == 'Gaza Strip'].event_location_district.value_counts()
df_district = pd.DataFrame({'area': district_based.index, 'count': district_based.values})

df_district.loc[~df_district['area'].isin(['Gaza', 'North Gaza', 'Khan Yunis']), 'area'] = 'Others'
df_district = df_district.groupby('area')['count'].sum().reset_index()

plt.style.use('_mpl-gallery-nogrid')

colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(df_district)))

# plot
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(df_district['count'], colors=colors, 
                                  autopct='%1.1f%%', radius=1, center=(0, 0),
                                  wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

ax.legend(wedges, df_district['area'], title='Areas', loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
ax.axis('equal')  

plt.show()
```

```python
# Means By which the people got killed
injuries = fatalities[['date_of_event', 'date_of_death', 'type_of_injury']]
def categorize_severity(row):
    if row['date_of_event'] == row['date_of_death']:
        return 'Same day'
    elif (row['date_of_death'] - row['date_of_event']).days <= 7:
        return 'Within week'
    else:
        return 'After 1 week'
        
injuries['severity'] = injuries.apply(categorize_severity, axis=1)
pivot_table = injuries.pivot_table(index='type_of_injury', columns='severity', aggfunc='size', fill_value=0)
pivot_table.columns = ['Same day', 'Within week', 'After 1 week']
pivot_table['Total'] = pivot_table.sum(axis=1)
pivot_table.reset_index(inplace=True)

# Display the resulting dataset
print(pivot_table)

# pie charts for better representation

plt.style.use('_mpl-gallery-nogrid')
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(pivot_table)))

# plot
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(pivot_table['Total'], colors=colors, 
                                  autopct='%1.1f%%', radius=1, center=(0, 0),
                                  wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

ax.legend(wedges, pivot_table['type_of_injury'], title='Injuries', loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
ax.axis('equal')  

plt.show()
```

```python
# Profile of victims

victims = fatalities[['name', 'age', 'citizenship', 'gender', 'took_part_in_the_hostilities', 'place_of_residence_district', 'event_location', 'date_of_death']]
victims['took_part_in_the_hostilities'].fillna('No', inplace=True)

victims
```

```python
# Identifyin common characteristics Based on their citizenship
grouped_citizenship = victims.groupby('citizenship').agg({
    'name': 'count',  
    'age': 'mean',    
})

grouped_citizenship.columns = ['victim_count', 'average_age']
grouped_citizenship.reset_index(inplace=True)
print(grouped_citizenship)
```

```python
#citizens getting killed based on their activity in taking part in hostilities
# Set the size of the displot
plt.figure(figsize=(12, 6))

# Your original code
sns.displot(data=victims, x="took_part_in_the_hostilities", col="citizenship", kde=True, height=6)

# Show the plot
plt.show()
```

```python
# Identifyin common characteristics Based on their residence
grouped_residence = victims.groupby('place_of_residence_district').agg({
    'name': 'count',  
    'age': 'mean',    
})

grouped_residence.columns = ['victim_count', 'average_age']
grouped_residence.reset_index(inplace=True)
print(grouped_residence)

fig = sns.catplot(data=victims, kind="strip", y="age", x="place_of_residence_district", hue="citizenship", dodge=True, alpha=0.5)
# Rotate x-axis labels by 90 degrees
fig.set_xticklabels(rotation=90)

# Show the plot
plt.show()
```

```python
# Identifyin common characteristics Based on the event location
grouped_location = victims.groupby('event_location').agg({
    'name': 'count',  
    'age': 'mean',    
})

grouped_location.columns = ['victim_count', 'average_age']
grouped_location.reset_index(inplace=True)
print(grouped_location)
```
